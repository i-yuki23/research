```python
import numpy as np
import matplotlib.pyplot as plt

for element in closet_water_distance_to_ligand_for_each_element:
    plt.figure()  # 新しい図を作成
    counts, bins, _ = plt.hist(closet_water_distance_to_ligand_for_each_element[element], bins=100, range=(0, 6), edgecolor='black')
    plt.xlabel('Distance')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {element}')
    
    # ピークのx座標を求める
    peak_index = np.argmax(counts)
    peak_x = (bins[peak_index] + bins[peak_index + 1]) / 2
    print(f'ピークのx座標 for {element}: {peak_x}')
    
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

element_to_radius = {'C': 1.69984, 'N': 1.62500, 'O': 1.51369, 'S': 1.78180, 'H': 1.2, 'B' : 1.92, 'F' : 1.47, 'P' : 1.80, 'I' : 1.98}

for element in closet_water_distance_to_ligand_for_each_element:
    embedding_dist = [(element_to_radius['O'] + element_to_radius[element]) - distance for distance in closet_water_distance_to_ligand_for_each_element[element]]
    plt.figure()  # 新しい図を作成
    counts, bins, _ = plt.hist(embedding_dist, bins=100, edgecolor='black')
    plt.xlabel('Overlapping Distance', fontsize=18)
    plt.ylabel('Frequency', fontsize=18)
    plt.title(f'Histogram of {element}', fontsize=18)
    plt.tick_params(axis='both', which='major', labelsize=16)
    
    # ピークのx座標を求める
    peak_index = np.argmax(counts)
    peak_x = (bins[peak_index] + bins[peak_index + 1]) / 2
    print(f'ピークのx座標 for {element}: {peak_x}')
    print(bins, counts)
    print(bins[-1])
    plt.show()
```
