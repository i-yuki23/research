```python
from modules.calculate_closest_water_to_ligand import calculate_closet_water_distance_to_ligand_for_each_element
from lib.pdb import get_all_pdb_names
import matplotlib.pyplot as plt
```

```python
closet_water_distance_to_ligand_for_each_element = calculate_closet_water_distance_to_ligand_for_each_element(get_all_pdb_names())
```

```python
for element in closet_water_distance_to_ligand_for_each_element:
    plt.figure()  # 
    plt.hist(closet_water_distance_to_ligand_for_each_element[element], bins=100, range=(0, 6), edgecolor='black')
    plt.xlabel('Distance')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {element}')
    plt.show()
```

```python
element_to_radius = {'C': 1.69984, 'N': 1.62500, 'O': 1.51369, 'S': 1.78180, 'H': 1.2, 'B' : 1.92, 'F' : 1.47, 'P' : 1.80, 'I' : 1.98}

for element in closet_water_distance_to_ligand_for_each_element:
    enbedding_dist = [(element_to_radius['O'] + element_to_radius[element]) - distance for distance in closet_water_distance_to_ligand_for_each_element[element]]
    plt.figure()  # 
    plt.hist(enbedding_dist, bins=100, edgecolor='black')
    plt.xlabel('Embedding Distance')
    plt.ylabel('Frequency')
    plt.title(f'{element}')
    plt.show()
```
